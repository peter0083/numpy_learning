```python
import numpy as np
import matplotlib.pyplot as plt
```

### random low and high temperature data

```python
data_set_size = 15 # this data set will have 15 data points
low_mu, low_sigma = 50, 4.3
low_data_set = low_mu + low_sigma * np.random.randn(data_set_size)
high_mu, high_sigma = 57, 5.2
high_data_set = high_mu + high_sigma * np.random.randn(data_set_size)

days = list(range(1, data_set_size +1)) # create a list which includes 
# a range from 1 to the size of data set
# this variable 'days' will be used for the x axis
```

```python
plt.plot(days, low_data_set)
plt.show()
```

```python
# now two lines in one plot
plt.plot(days, low_data_set,
         days, high_data_set)
plt.show()
```

```python
# improve the visibility and readiblity of this plot
plt.plot(days, low_data_set,
         days, low_data_set, "vm",
         days, high_data_set,
         days, high_data_set, "^k")
# mark the data points of low_data_set by downward pointing arrows using 'v'
# 'm' indicates the colour 'magenta'
# mark the data points of high_data_set by upward pointing arrows using '^'
# 'k' indicates the colour 'black'
plt.show()
# this plot actually has four data sets (2 low_data_set and 2 high_data_set)
```

```python
plt.plot(days, low_data_set,
        
         days, high_data_set
        )
plt.xlabel('Day')
plt.ylabel('Temperature: degrees Farenheit')
plt.title('Randomized temperature data')
plt.show()
```

```python
plt.plot(
         days, high_data_set, "^k") # plot only using the upward arrows
plt.xlabel('Day')
plt.ylabel('Temperature: degrees Farenheit')
plt.title('Randomized temperature data')
plt.show()
```

Lynda.com adapted the example below from: https://matplotlib.org/examples/pylab_examples/legend_demo2.html

```python
# create two time series
# the range of generated values starts with the first parameter and continue up to 
# but not including the second parameter in increments equal to the third parameter
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)
# thus, we have ten times more data points in t2 than t1

# note that plot returns a list of lines. The "L1, = plot" usage
# extracts the first element of the list into L1 using tuple
# unpacking. So L1 is a Line2D instance, not a sequence of lines
l1, =plt.plot(t2, np.exp(-t2))
l2, l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--go', t1, np.log(1 + t1), '.')
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 'rs-.')

plt.legend((l2, l4), ('oscillatory', 'damped'), loc='upper right', shadow=True)
# first parameter: a tuple f lines that we want to use in the legend
# second parameter: a tuple of strings for the legend
# third parameter: name, indicates the location of the legend
# fourth parameter: name, indicates styling for the legend
plt.xlabel('time')
plt.ylabel('volts')
plt.title('Damped oscillation')
plt.show()
```
