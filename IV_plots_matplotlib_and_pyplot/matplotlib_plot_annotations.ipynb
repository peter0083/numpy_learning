```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_data_points = 10

my_figure = plt.figure() # create a figure
subplot_1 = my_figure.add_subplot(1, 1, 1) # create a subplot
subplot_1.plot(np.random.rand(number_of_data_points).cumsum()) # add randomized data to the subplot

subplot_1.text(1, 0.5, r'an equation: $E=mc^2$', fontsize=18, color='red')
# (1, 0.5) ---> coordinates of where the annotated text is shown
# x coordinate 1 is given in terms of the number of data points we're displaying
# y coordinate 0.5 is given using the y axis as a reference point.
# r'' indicates a raw string -> backslashes wil not be interpreted as escape characters
# r'' is very handy to prepend whenever your string includes equations

subplot_1.text(1, 1.5, 'Hello, Mountain Climbing!', 
               family='monospace', 
               fontsize=14, 
               color='green')

# see: http://matplotlib.org/users/transforms_tutorial.html
# transform=subplot_1.transAxes; entire axis between zero and one
subplot_1.text(0.5, 0.5, "We are centered now", transform=subplot_1.transAxes)
# this transform function transform both x and y axis from whatever the 
# data range is to zero-to-one. 
# so this coordinate (0.5, 0.5) means center of the x axis and center of the y axis
# aka center of the plot

# arrow annotation
subplot_1.annotate('shoot arrow', xy=(2, 1), xytext=(3, 4),
                 arrowprops=dict(facecolor='red', shrink=0.05))
# xy=(2, 1) indicates where the arrow is pointing
# xytext=(3, 4) indicates where the text begins

plt.show()
```

```python
# dampened sine wave plot

x = np.arange(0, 10, 0.005)
y = np.exp(-x/2.) * np.sin(2*np.pi*x)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

xdata, ydata = 5, 0
xdisplay, ydisplay = ax.transData.transform_point((xdata, ydata))

# bbox and boxstyle for the annotation
# to make the plot for readible and usable
bbox = dict(boxstyle="round", fc="0.8")
# used a dictionary wth box style
arrowprops = dict(
    arrowstyle = "->",
    connectionstyle = "angle,angleA=0,angleB=90,rad=10")
# used dictionary with arrow props

offset = 72
ax.annotate('data = (%.1f, %.1f)'%(xdata, ydata),
            (xdata, ydata), 
            xytext=(-2*offset, offset),
            textcoords='offset points',
            bbox=bbox,
            arrowprops=arrowprops)

disp = ax.annotate('display = (%.1f, %.1f)'%(xdisplay, ydisplay),
                  (xdisplay, ydisplay), 
                  xytext=(0.5*offset, -offset),
                  xycoords='figure pixels',
                  textcoords='offset points',
                  bbox=bbox,
                  arrowprops=arrowprops)

plt.show()
```

```python
fig = plt.figure()
for i, label in enumerate(('A', 'B', 'C', 'D')): # use the loop to position each subplot
    # each subplot is labeled to show the position
    ax = fig.add_subplot(2,2,i+1)
    ax.text(0.05, 0.95, 
            label, transform=ax.transAxes,
            fontsize=16, 
            fontweight='bold',
            va='top')
    
plt.show()
```
