```python
import numpy as np
import matplotlib.pyplot as plt
```

this is an example from: http://matplotlib.org/examples/pie_and_polar_charts/pie_demo_features.html

In additionto a basic pie chart, this demo shows a few optional features

- slide labels
- auto-labelling the percentage
- ofsetting a slice with "explode"
- drop-shadow
- custom start angle

**Warning: Unless you have a very very good reason, avoid pie chart and use bar chart **
Viz expert Dr. Tamara Munzner suggests that bar chart is way more effective than pie chart.
For usability and readibility, stick with bar chart!

Dr. Tamara Munzer's talk on visualization principles: https://www.youtube.com/watch?v=ZgOF8R6YL2U

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0) # only explode the 2nd slice (ie. Hogs)

plt.pie(x=sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors,
        autopct='%1.1f%%', 
        # indicates the display format Percentage using python string formatting. 
        # the floating point values will be displayed wth one numeral after the decimal point
        shadow=True,
        startangle=90)
# how to use %1.1f%%
# https://stackoverflow.com/questions/6170246/how-do-i-use-matplotlib-autopct
# https://www.learnpython.org/en/String_Formatting

# set aspect ratio to be equal so that pie is drawn as a circle
plt.axis('equal')

plt.show()
```

### Bar Chart
This exampe is taken from: http:/matplotlib.org/examples/api/barchart_demo.html

API location: http://matplotlib.org/api/pyplot_api.html

matplotlib.pyplot.bar(left, height, width=0.8, bottom=None, hold=None, data=None, **kwargs)

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
menStd = (2, 3, 4, 1, 2)

ind = np.arange(N) # the x locations for the groups
# ind means independent variables
width = 0.35 # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, menMeans, width, color='r', yerr=menStd)
# yerr is where we specify the standard deviation 

womenMeans = (25, 32, 34, 20, 25)
womenStd = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, womenMeans, width, color='y', yerr=womenStd)

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')

ax.set_xticks(ind + width)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))

plt.show()
```
