```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 100, 15 # Python allows us to assign multiple variables in one line
data_set = mu + sigma * np.random.randn(10000) # generate a dataset with 10,000 data points
# with a mean of 100 and a standard deviation of 15

# the histogram of the data
n, bins, patches = plt.hist(data_set, 50, normed=1, facecolor='g', alpha=0.75)
# bins = 50
# value of the data points are normalized to 1. that is, their sum is equal to one
# colour is green
# alpha transparency is 0.75

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.025, r'$\mu-100,\ \sigma-15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
